### График, который показывает, как меняется приближенное значение площади в зависимости от количества сгенерированных точек

```python
import matplotlib.pyplot as plt

# возможно тут надо поменять название файла (график построен для нужного)
file_name = "A1_Different_number_of_points_1.txt"

# Чтение файла
with open(file_name, "r") as file:
    lines = file.readlines()
    x = list(map(float, lines[0].strip().split(',')))
    y = list(map(float, lines[1].strip().split(',')))

# Построение графика
plt.figure(figsize=(15, 6))
plt.plot(x, y, marker='o', linestyle='-', color='b', label="Вычисленная площадь")

# Горизонтальная прямая - значение точной площади
plt.axhline(y=0.944517, color='green', linestyle='-', label="Точная площадь")

# Подписи
plt.xlabel("Количество точек")
plt.ylabel("Вычисленная площадь фигуры")
plt.title("График функции")
plt.legend()
plt.grid(True)

plt.show()
```

- Видим, что чем больше точек, тем более точно считается площадь (то есть с увеличением числа точек площадь стабилизируется около некоторого наперёд заданного числа - точной площади кусочка)

### График, который показывает, как меняется приближенное значение площади в зависимости от размера рассматриваемой области (прямоугольника, количество сгенерированных точек равно 100000)

Первый  рассматриваемый прямоугольник - наименьший возможный прямоугольник, в который будет помещаться пересечение трёх окружностей целиком. Далее постепенно увеличиваем его - все границы увеличиваем на некоторую delta на каждом шаге, вычисляем площадь и записываем в массив. Количество точек было решено взять равным 100000, чтобы точнее определить корреляцию между масштабом и точностью вычисленной площади - во-первых, в таком случае колчиество точек всегда одинаковое и не повлияет на ответ; во-вторых, как было сказано в прошлом выводе, при увеличении числа точек увеличивается точность нахождения площади. Поэтому при числе точек равном 100000 (максимально возможном) мы получим хорошую функцию зависимости точности площади от масштаба

```python
# возможно тут надо поменять название файла (график построен для нужного)
file_name = "A1_Different_Scale_1.txt"

# Чтение файла
with open(file_name, "r") as file:
    lines = file.readlines()
    x = list(map(float, lines[0].strip().split(',')))
    y = list(map(float, lines[1].strip().split(',')))

# Построение графика
plt.figure(figsize=(15, 6))
plt.plot(x, y, marker='o', linestyle='-', color='b', label="Вычисленная площадь")

# Горизонтальная прямая - значение точной площади
plt.axhline(y=0.944517, color='green', linestyle='-', label="Точная площадь")

# Подписи
plt.xlabel("delta")
plt.ylabel("Вычисленная площадь фигуры")
plt.title("График функции")
plt.legend()
plt.grid(True)

plt.show()
```

- Видим, что чем меньше рассматриваемая область, тем в среднем более точно считается площадь. Это логично, поскольку в меньшей области будет "меньше" точек, а значит будет не такой большой разброс

### График, который показывает, как меняется величина относительного отклонения приближенного значения площади от ее точной оценки в зависимости от числа сгенерированных точек
